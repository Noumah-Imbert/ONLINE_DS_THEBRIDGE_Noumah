### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python

 
df.head(10) #cambie el indice por la columna city
top5_values_indices = df.nlargest(5, 'population_total')

colores = ['blue', 'grey', 'red','olive','brown']
transparencia = [0.4, 0.4, 0.9, 0.4, 0.4]
fig, ax = plt.subplots()

for i in range(len(top5_values_indices)):
    bar = ax.bar(top5_values_indices.index[i], top5_values_indices['population_total'][i], color=colores[i], alpha=transparencia[i])
    ax.text(bar[0].get_x() + bar[0].get_width() / 2, bar[0].get_height(), str(top5_values_indices['population_total'][i]),
            ha='center', va='bottom')
plt.xlabel('Ciudades')
plt.ylabel('Habitantes(miles)')
plt.title('Ciudades mas pobladas de California')
```

```python
#2
''' 0   Unnamed: 0          427 non-null    int64  
 1   city                427 non-null    object 
 10  area_total_km2      427 non-null    float64
histograma de area ocuapda por ciudades 
 (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
 '''
fuente = {'family': 'Bradley Hand ITC', 'size': 18, 'color': 'black'}
fig,ax = plt.subplots()
plt.hist(df['area_total_km2'], bins=50, color='steelblue', alpha= 0.2, label = "Hist Area")
plt.xlabel('Superficie ($km^2$)', fontdict=fuente)
plt.ylabel('Numero de Ciudades', fontdict=fuente)
plt.legend(loc='upper right');
```
