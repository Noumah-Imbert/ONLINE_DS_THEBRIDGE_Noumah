```python
df3= df.copy()
df3 = df3.iloc[:, :24]
columns_to_drop = ['Adjusted Grade', 'New?', 'Other Location Code in LCGMS']
df3 = df3.drop(columns=columns_to_drop)
df3.info()
df3
```

```python
#Analizar geogrficamente las zonas de NY, en funcin de las necesidades econmicas y las estimaciones de ingresos de los colegios
indice = df.groupby("District")[["Economic Need Index"]].mean().reset_index().sort_values(by= 'District', ascending= False)

df['School Income Estimate'] = df['School Income Estimate'].replace('[\$,]', '', regex=True).astype(float)

# Sumar los valores de la columna 'School Income Estimate' por distrito
suma_por_distrito = df.groupby('District')[['School Income Estimate']].mean().reset_index().sort_values(by= 'District', ascending= False)

suma_por_distrito
```

```python
indice
```

```python
###### Comparacion de la necesidad (promedio) de las escuelas de cada distrito y de sus ingresos (promedios)
resultado = pd.merge(suma_por_distrito, indice, on='District')
resultado["School Income Estimate"] = resultado["School Income Estimate"].round(2)
resultado["Economic Need Index"] = resultado["Economic Need Index"].round(2)

resultado
```

```python

#2. Cmo se reparten geogrficamente los colegios en funcin de la raza de sus estudiantes?
# Configuracin de estilo de Seaborn
sns.set(style="whitegrid")

# Lista de columnas de porcentajes
columnas_porcentajes = ['Percent ELL', 'Percent Asian', 'Percent Black', 'Percent Hispanic', 'Percent Black / Hispanic', 'Percent White']

# Visualizacin de barras para cada columna
for columna in columnas_porcentajes:
    plt.figure(figsize=(15, 10))

    
    sns.barplot(x='City', y=columna, data=df)
    plt.title(f'Distribucin de estudiantes por {columna} en diferentes ciudades')
    plt.xlabel('Ciudad')
    plt.ylabel(f'Porcentaje de estudiantes {columna}')
    plt.xticks(rotation=45, ha='right')  
```

```python
def plot_distribution_by_city(df, percentage_column):
    df_filtered = df3[(df3[percentage_column].notna()) & (df3[percentage_column] != 0)]


    # Crea un grfico de tarta con Seaborn
    plt.figure(figsize=(10, 8))
    sns.set(style="whitegrid")
    sns.set_palette("pastel")
    sns.set_color_codes("pastel")

    # Conteo de colegios por ciudad
    count_by_city = df_filtered['City'].value_counts()

    # Grfico de tarta
    plt.pie(count_by_city, labels=count_by_city.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribucin de colegios con estudiantes en la columna {percentage_column} en Nueva York')
    plt.show()


```

```python
#2.1 En qu zonas de NY se encuentran la mayor parte de colegios con estudiantes de raza negra?

plot_distribution_by_city(df, 'Percent Black')

```

```python
#2.2 Lo mismo para raza hispana
plot_distribution_by_city(df, 'Percent Hispanic')
```

```python
#2.3 Raza blanca
plot_distribution_by_city(df, 'Percent White')
```

```python
#2.4 Asitica

plot_distribution_by_city(df, 'Percent Asian')
```
